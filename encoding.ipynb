```python
cleaned_df = cleaned_df.drop(["COVID_HR1","COVID_HR2","IDEOLOGY4"],axis=1)
missing = ["HR11","IDEOLOGY7"]
for i in missing:
    mode = cleaned_df[i].mode()
    print(mode[0])
    fixed = []
    for value in cleaned_df[i]:
        if value == 'NA' or value != value:
            fixed.append(mode[0])
        else:
            fixed.append(value)
    cleaned_df[i] = fixed
```

### Chi Squared Feature Selection

```python
#A-L
tmp_df = cleaned_df.copy()
#tmp_df = tmp_df.drop("HR11",axis=1)

tmp_df = tmp_df.drop("IDEOLOGY5",axis=1)
nominal_val = cleaned_df.select_dtypes(include=['object']).columns
for index, values in tmp_df.iteritems():
    if index in nominal_val:
        unique = pd.unique(values)
        dic = []
        for i in values:
            print(np.where(unique == i))
            print(index)
            dic.append(np.where(unique == i)[0][0])
        tmp_df[index] = dic
```

```python
new_df = cleaned_df.copy().drop("IDEOLOGY5",axis = 1)
from sklearn.model_selection import train_test_split
test = new_df.columns
target_feature = []
for i in test:
    if i == "HR5":
        target_feature.append(i)
X = tmp_df.drop("HR5",axis=1) #obamacare
y = tmp_df["HR5"]
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = .75, test_size = .25, random_state = 72)
```

```python
from sklearn.feature_selection import SelectKBest, chi2
fs = SelectKBest(score_func=chi2, k='all')
fs.fit(x_train,y_train)
x_train_fs = fs.transform(x_train)
x_test_fs = fs.transform(x_test)
```

```python
from matplotlib import pyplot
scores = []
selected_features = []
for i in range(len(fs.scores_)):
    #rint('Feature %s: %f' % (new_df.columns[i], fs.scores_[i]))
    scores.append(fs.scores_[i])
    if fs.scores_[i] > 80:
        selected_features.append(new_df.columns[i])
# plot the scores
# pyplot.bar([i for i in range(len(fs.scores_))], fs.scores_)
# pyplot.show()
```
